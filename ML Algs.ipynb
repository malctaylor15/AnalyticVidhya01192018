For missing values, we can 
1. Look for all complete rows
2. Impute missing value indicator 999/ 99999 (might be better for trees) 
3. Impute mean (for linear regression) 

```python
fill_nas = modeling_train.apply(lambda x: x.mean(skipna=True), axis=0)
```

```python
demo = modeling_train.apply(lambda x: x.fillna(x.mean()), axis=0)
```

## Start ML Process

```python
X = demo.drop(["Approved"], axis = 1)
Y = demo["Approved"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.25, random_state = 1234) # , shuffle = True)
```

```python
from sklearn.metrics import roc_auc_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
def log_metrics(true_y, scores):
    train_roc = roc_auc_score(true_y, scores)
    train_acc =  accuracy_score(true_y, scores)

    print("--- Confusion Matrix ---")
    print(confusion_matrix(true_y, scores))
    print("Train ROC: ", train_roc) 
    print("Accuracy: ", 100*train_acc,"%")
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
lin_fit = LogisticRegression(fit_intercept=True, class_weight={1:100, 0:1}).fit(X_train, Y_train)
```

```python
lin_train_scores = lin_fit.predict(X_train)
lin_test_scores = lin_fit.predict(X_test)

log_metrics(Y_train, lin_train_scores)
log_metrics(Y_test, lin_test_scores)
```

### Random Forest 

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf1 = RandomForestClassifier(class_weight={1:100, 0:1}, n_estimators=50, verbose=1, max_depth=7).fit(X_train, Y_train)
```

```python
rf1_train_scores = rf1.predict(X_train)
rf1_test_scores = rf1.predict(X_test)

log_metrics(Y_train, rf1_train_scores)
log_metrics(Y_test, rf1_test_scores)
```

### Grid Search Random Forest 

```python
from sklearn.model_selection import RandomizedSearchCV, GridSearchCV
```

```python
np.random.seed(1234)
param_dist_rf = {"n_estimators": [50, 100, 200], 
              "max_depth": [3, 6, 9, None],
              "max_features": ["auto", None, 3, 7, 12]
              }
```

```python
rf_random_search = GridSearchCV(RandomForestClassifier(verbose = 1, class_weight={1:100, 0:1}), 
                                param_grid=param_dist_rf, cv = 3, n_jobs=3)
```

```python
rf_random_search.fit(X, Y) # Use full set because are doing CV as the overfit check 
```

```python
wanted_grid_params = ["mean_train_score" ,"mean_test_score", "param_n_estimators", "param_max_depth", 
                      "param_max_features", "rank_test_score"]
```

```python
dict_temp = {x:rf_random_search.cv_results_[x] for x in wanted_grid_params}
search_results = pd.DataFrame(dict_temp)
search_results["test_train_pct_diff"] = (search_results.mean_train_score - search_results.mean_test_score)/ search_results.mean_test_score
search_results[search_results.test_train_pct_diff < 0.10]\
.sort_values(by=["mean_test_score"], ascending = False).head(15)
```

*Look for best parameters in Grid * 

```python
rf_better = RandomForestClassifier(verbose = 1, class_weight={1:100, 0:1},
                                   n_estimators = 90, max_depth = 9, max_features = 'auto')
```

```python
rf_better.fit(X_train, Y_train)
```

```python
rf_better_train_scr = rf_better.predict(X_train)
rf_better_test_scr = rf_better.predict(X_test)

log_metrics(Y_train, rf_better_train_scr)
log_metrics(Y_test, rf_better_test_scr)
```

Make sure it is not over fitting and that the parameters are the best 

```python
rf_best = rf_better.fit(X,Y)
```
